```python
api_key='8265bd1679663a7ea12ac168da84d2e8'
import requests
import datetime
import time
import pandas as pd
print("COLLECTING INDIAN(hindi) MOVIES")
start_time = datetime.datetime.now()
final_list =[]
n=0
while n<296:
    n+=1
    url = "https://api.themoviedb.org/3/discover/movie?&with_origin_country=IN&&with_original_language=hi&page={}&api_key=".format(n)
    res = requests.get(url+api_key).json()
    data = res['results']
    final_list.extend(data)
    time.sleep(2)
    if n%5 == 0:
        print(f'{n} pages completed')
print("----------------")
print("MOVIES COLLECTED")
print("----------------")
size = len(final_list)
print(f"TOTAL NUMBER OF MOVIES COLLECTED {size}")
print("----------------------------------------")
print("CREATING DATAFRAME")
print("------------------")
indian_movies_data = pd.DataFrame(final_list)
print("DATAFRAME CREATED")
print("-----------------")
print("COLLECTING KEYWORDS FOR EACH MOVIE")
print("-----------------------------------")
KEYWORDS = []
c=0
for i in indian_movies_data.id:
    sid=i
    c+=1
    info = movie.details(sid)
    movie_keywords = []
    for j in range(len(info.keywords.keywords)):
        g = info.keywords.keywords[j].name
        movie_keywords.append(g)
    KEYWORDS.append(movie_keywords)
    if c%50 == 0:
        print(f'KEYWORDS COLLECTED OF {c} MOVIES')
print("ALL KEYWORDS COLLECTED")
print("-----------------------")
print("MERGING KEYWORDS")
print("-----------------")
indian_movies_data['KEYWORDS'] = KEYWORDS
print("COLLECTING GENRES FOR EACH MOVIE")
print("--------------------------------")
GENRES = []
c=0
for i in indian_movies_data.id:
    sid=i
    c+=1
    info = movie.details(sid)
    movie_genre = []
    for j in range(len(info.genres)):
        g = info.genres[j].name
        movie_genre.append(g)
    GENRES.append(movie_genre)
    if c%50 == 0:
        print(f'GENRES COLLECTED OF {c} MOVIES')
print("ALL GENRES COLLECTED")
print("--------------")
print("MERGING GENRES")
print("---------------")
indian_movies_data['GENRES'] = GENRES
print("COLLECTING CAST FOR EACH MOVIE")
print("------------------------------")
all_cast = []
c=0
for i in indian_movies_data.id:
    sid=i
    c+=1
    cred = requests.get("https://api.themoviedb.org/3/movie/{}?api_key=8265bd1679663a7ea12ac168da84d2e8&append_to_response=credits".format(sid)).json()
    actors_name = []
    for j in range(len(cred['credits']['cast'])):
        n = cred['credits']['cast'][j]['name']
        actors_name.append(n)
    all_cast.append(actors_name)
    if c%50 == 0:
        print(f'NUMBER OF MOVIES COMPLETED: {c}')
print("ALL CAST COLLECTED")
print("--------------")
print("MERGING CAST")
print("------------")
indian_movies_data['CAST'] = all_cast
print("DATA FETCHING COMPLETED")
print("-----------------------")
print("CREATING CSV")
indian_movies_data.to_csv('indian_movies_data.csv', header=True, index=False)
print("------------")
print("ACCESS DATA BY indian_movies_data AND CSV BY indian_movies_data.csv")
end_time = datetime.datetime.now()
print('TOTAL TIME TAKEN {}'.format(end_time-start_time))
```
