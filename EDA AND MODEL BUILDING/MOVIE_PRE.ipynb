```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import ast
```

```python
main_df = pd.read_csv('../MOVIE_RECOMMENDATION/DATA/indian_movies_data_director.csv')
main_df.head()
```

```python
main_df = main_df.drop(['adult','backdrop_path','genre_ids','original_language','original_title','video'],axis=1)
```

```python
main_df = main_df.dropna()
```

```python
main_df = main_df.reset_index()
```

```python
main_df = main_df.drop(['index','poster_path'],axis=1)
```

```python
main_df
```

```python
def converter(obj):
    l = []
    for i in ast.literal_eval(obj):
        l.append(i)
    return l
```

```python
main_df['GENRES'] = main_df['GENRES'].apply(converter)
main_df['CAST'] = main_df['CAST'].apply(converter)
main_df['KEYWORDS'] = main_df['KEYWORDS'].apply(converter)
main_df['DIRECTORS'] = main_df['DIRECTORS'].apply(converter)
```

```python
main_df
```

```python
#main_df.to_csv('indian_movies_usefull1.csv', header=True, index=False)
```

```python
main_df = main_df[main_df.vote_count > 10]
main_df = main_df[main_df.popularity > 2]
main_df = main_df[main_df.release_date > '2010-01-01']
```

```python
main_df
```

```python
main_cast = []
for i in main_df.CAST:
    name = []
    name.append(i[0])
    if len(i)>1:
        name.append(i[1])
    main_cast.append(name)
```

```python
main_cast
```

```python
main_df['lead'] = main_cast
```

```python
main_df
```

```python
df = main_df
df['overview'] = df['overview'].apply(lambda x:x.split())
df['GENRES'] = df['GENRES'].apply(lambda x:[i.replace(" ","") for i in x])
df['CAST'] = df['CAST'].apply(lambda x:[i.replace(" ","") for i in x])
df['KEYWORDS'] = df['KEYWORDS'].apply(lambda x:[i.replace(" ","") for i in x])
df['DIRECTORS'] = df['DIRECTORS'].apply(lambda x:[i.replace(" ","") for i in x])
df['lead'] = df['lead'].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
df['tags'] =  df['lead'] + df['KEYWORDS'] +  df['GENRES']  + df['DIRECTORS']# + df['overview']
```

```python
new_df = df[['id','title','tags']]
```

```python
new_df
```

```python
new_df['tags'] = new_df['tags'].apply(lambda x:" ".join(x))
new_df['tags'] = new_df['tags'].apply(lambda x:x.lower())
```

```python
import nltk
from nltk.stem.porter import PorterStemmer
from nltk.stem import WordNetLemmatizer
ps = PorterStemmer()
lemmatizer = WordNetLemmatizer()


def stem(text):
    y = []
    for i in text.split():
        y.append(ps.stem(i))
        
    return " ".join(y)
```

```python
new_df['tags'] = new_df['tags'].apply(stem)
```

```python
new_df['tags'][0]
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000,stop_words='english')
vector = cv.fit_transform(new_df['tags']).toarray()
vector.shape
```

```python
from sklearn.metrics.pairwise import cosine_similarity
similarity = cosine_similarity(vector)
similarity
```

```python
def recommend(movie):
    index = new_df[new_df['title'] == movie].index[0]
    distances = sorted(list(enumerate(similarity[index])),reverse=True,key = lambda x: x[1])
    for i in distances[1:11]:
        print(new_df.iloc[i[0]].title)
```

```python
recommend('Bajirao Mastani')
```
